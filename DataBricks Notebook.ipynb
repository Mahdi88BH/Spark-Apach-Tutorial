# Magic commands

1. **%md (Markdown)**
- This is used for documentation. It turns a cell into a text editor where you can use Markdown syntax to add headings, images, lists, and links.
- Utility: Making your notebook readable for others by explaining your logic or documenting results.

2. **%fs (File System)**
- This allows you to access Databricks File System (DBFS) directly. It’s a wrapper for the dbutils.fs utility.
- Utility: Listing files (ls), moving files (mv), or deleting them (rm) without needing to write complex Python code.
  - Example: **%fs ls /databricks-datasets**

3. **%sh (Shell)**
- This command lets you run shell commands on the driver node of your cluster.
- Utility: Installing specific libraries via apt-get, checking disk space, or interacting with the underlying Linux operating system.
  - Example: **%sh df -h**

4. **%run**
- This is used to import another notebook into your current one.
- Utility: Code modularity. If you have a notebook full of helper functions or configuration variables, you can %run it at the start of your main notebook to make all those functions available.
  - Example: **%run ./Shared/Common_Functions**

5. **%sql**
- This tells the notebook to interpret the cell as Spark SQL.
- Utility: Querying tables or temporary views directly. It’s often used by data analysts who prefer SQL over Python for data exploration.
  - Example: **SELECT * FROM sales_data WHERE amount > 100**

6. **%python**
- This forces a cell to run as Python code.
- Utility: Since Databricks notebooks usually have a "default language," you use this if your notebook is set to SQL or Scala but you need to write a quick Python script.
